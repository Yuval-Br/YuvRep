```python
import pandas as pd 
import matplotlib.pyplot as plt


djan = pd.read_csv(r'/content/Sales_January_2019.csv')
djan['Month'] = 1
dfeb = pd.read_csv(r'/content/Sales_February_2019.csv')
dfeb['Month'] = 2
dmar = pd.read_csv(r'/content/Sales_March_2019.csv')
dmar['Month'] = 3
dapr = pd.read_csv(r'/content/Sales_April_2019.csv')
dapr['Month'] = 4
dmay = pd.read_csv(r'/content/Sales_May_2019.csv')
dmay['Month'] = 5
djun = pd.read_csv(r'/content/Sales_June_2019.csv')
djun['Month'] = 6
djul = pd.read_csv(r'/content/Sales_July_2019.csv')
djul['Month'] = 7
daug = pd.read_csv(r'/content/Sales_August_2019.csv')
daug['Month'] = 8
dsep = pd.read_csv(r'/content/Sales_September_2019.csv')
dsep['Month'] = 9
doct = pd.read_csv(r'/content/Sales_October_2019.csv')
doct['Month'] = 10
dnov = pd.read_csv(r'/content/Sales_November_2019.csv')
dnov['Month'] = 11
ddec = pd.read_csv(r'/content/Sales_December_2019.csv')
ddec['Month'] = 12

totd = pd.concat([djan,dfeb,dmar,dapr,dmay,djun,djul,daug,dsep,doct,dnov,ddec]) # merging all together
#totd.to_excel('complete sales data.xlsx',index=False) #saving
totd = totd[totd['Product'] != 'Product'] # removing the unwanted rows.

totd['Quantity Ordered'] = pd.to_numeric(totd['Quantity Ordered'])
totd['Price Each'] = pd.to_numeric(totd['Price Each'])
totd['Sales'] = totd['Quantity Ordered'] * totd['Price Each']  # tot sales col

res = totd.groupby(['Month']).sum()
res1 = res.sort_values('Sales',ascending=False) # the month with the best sales at the top.
res2 = res.sort_values('Month')
months = range(1,13)
#plt.bar(months,res2['Sales'])
#plt.xlabel('month',fontsize=15)
#plt.ylabel('sales in thousend usd',fontsize=15)

totd['Purchase Address'] = totd['Purchase Address'].astype(str)  
totd['City'] = totd['Purchase Address'].apply(lambda x: x.split(',')) # spliting the strings in purschase address
totd['c'] = totd['City'].apply(lambda x: len(x))  
totd = totd.drop(totd[totd.c < 3].index) # keeping all the right ones (removing nans)
totd['City'] = totd['City'].apply(lambda x: x[1])  # taking just the city part (between the , )
totd = totd.drop(columns=['c'])
# apply() is used for applying functions on columns elementwise.
ans = totd.groupby(['City']).sum()
cities = pd.unique(totd['City'].sort_values())
#plt.bar(cities,ans['Sales'])
#plt.xticks(cities,rotation='vertical')

totd['Order Date'] = pd.to_datetime(totd['Order Date'])  # change to date time format
totd['Hour'] = totd['Order Date'].dt.hour # creating an order hour column
```
