```
 # recursive function example :

 def fib(n) :    # the n fibonacci number
   if n == 0 or n == 1 :
     return 1
   else :
        return fib(n-1) + fib(n-2)

def fibsum(n) :
   if n == 0 :
     return 1
   else :
     return fib(n) + fibsum(n-1)

 print(fib(5),fibsum(4))    
```

```
# integration :
import numpy as np
import scipy.integrate as integrate 

# standard intagral : 
f1 = lambda x : np.cos(np.exp(x))  # anonymous function of x

res1 = integrate.quad(f1, 0, 1) # the integral from 0 to 1 of f1. returns value and error estimate
print(res1)

print('---------')

# with parameters :

def f2(a,b,x) :
  return a*np.sqrt(x+b)   # function with pars

def res2(a,b,c,d) :
  return integrate.quad(lambda x : f2(a,b,x), c, d) # integrates f2 with chosen pars in chosen bounds

print(res2(2,6,1,5)) # example

print('---------')

# double integral :
 
f3 = lambda x,y : x**2 + y**3 + x*y
res3 = integrate.dblquad(f3, 0, 1, 1, 2)  # integrates f3 over  0<x<1 , 1<y<2  
print(res3)

print('---------')

# samples integration, given samples and not the whole function :

x = np.array([0, 0.25, 0.5, 0.75, 1])
y = np.array([2, 4.11, 5.2, 6.77, 3])

res11 = integrate.simps(y,x)  # two ways of approximating with samples from f
res22 = integrate.trapz(y,x)
print(res11, res22)




```

```
# two variable optimization : 

import numpy as np 
from scipy import optimize
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

def f1(x,y) :
  return x**4 + y**2 - 10*x*y
x = np.linspace(-15, 15, 100)               # plotting a surface
y = np.linspace(-10, 20, 100)
X, Y = np.meshgrid(x, y)     # the cordiniate matrix of x,y
Z = f1(X, Y)                  # f1 of it. the surface equation
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)

def f(u) :
  f = u[0]**4 + u[1]**2 - 10*u[0]*u[1]
  return f

def jf(u) :
  df = np.array([4*u[0]**3 - 10*u[1], 2*u[1] - 10*u[0]])
  return df

# min no constraints no jac :
print(optimize.minimize(f, np.array([5, 15]), method='nelder-mead'))  # takes function and initial guess
print('---------')

# min no constraints yes jac :
print(optimize.minimize(f, np.array([5, 15]), method='BFGS', jac=jf))   # faster with jacobian
print('----------')

# min with conastraints :
ineq_cons = {'type': 'ineq','fun' : lambda x : x[1]-x[0], 'jac' : lambda x : np.array([-1,1])}   # fun>0 and its jac
eq_cons = {'type': 'eq', 'fun' : lambda x : x[0]**2-x[1], 'jac' : lambda x : np.array([2*x[0], -1])}  # fun=0 and its jac
print(optimize.minimize(f,np.array([1,0]),method='SLSQP',jac=jf,constraints=[eq_cons,ineq_cons]))
print('----------')

# global min with bounds :
b = [(-1,1),(2,10)]  
print(optimize.shgo(f,b))
print('----------')
print(optimize.dual_annealing(f,b))
print('----------')


# numerical jacobian :
def numeric_jf(u) :
  h = 10**(-4)   # the small derivitive h
  ndf = np.array([(f(u+np.array([h,0]))-f(u))/h,(f(u+np.array([0,h]))-f(u))/h])  # der approx
  return ndf
print(optimize.minimize(f, np.array([5, 15]), method='BFGS', jac=numeric_jf ))   # can compere yes/no jac and num/reguler jac
print('----------')

# for function max you can do the same procces for -f and take the x value and -fun value

```

```
#  linalg uses :
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg


# least squares linear regression :

x = np.array([[1.5,2.3,3,3.5,3.9,4.2,5,5.5,6]])             # set of x values
y = np.array([[2, 3.1, 4.4, 6, 10, 16.9, 21, 25.33, 28]])              # set of y values
x1 = np.log(x)                               # the target function is y =a*log(x)+b*x**2
x2 = x**2 
s = np.concatenate((x1.T, x2.T), axis=1)                 
sol = linalg.lstsq(s, y.T)                       # solves it with least squares
a, b = sol[0]              # the coefficients
xx = np.linspace(1.5,6,50)
yy = a*np.log(xx) + b*xx**2
plt.plot(x[0],y[0],'o',label='original samples')         # label once. 'o' so it will apear in dots without lines
plt.plot(x[1:],y[1:],'o')
plt.plot(xx,yy,label='fitted curve')
plt.legend()
plt.xlabel('time',fontsize=15,color='r')              #axes labels
plt.ylabel('population',fontsize=15,color ='r')

print('-------------')

# linear eq system :  m*u=v

# if m is invertible :

m0 = np.array([[1.2,5,4.6],[2.2,3.4,6.68],[3,4.56,6.78]])  # invertible 3X3 matrix
v0 = np.array([1, 1.23, 2.34]) 
m00 = linalg.inv(m0)
u01 = np.dot(m00,v0.T) # u0 = inv(m0)*v0
u02 = linalg.solve(m0,v0.T)  # better
print(u01,u02)
print(np.dot(m0,u02))  # exact solution

print('-----------')

# more general : pinv

m1 = np.array([[1.5, 1.79, 2.45],[1.35, 4.11, 2.5],
               [1.52 ,1, 2.9],[1, 3.55, 2.77]])            # 4X3 matrix !

v1 = np.array([1.45, 2.33, 1.56, 4.4])  # 4X1 sol column
m11 = linalg.pinv(m1)    # a psuedo inverse with least squares. 3x4
u1 = np.dot(m11,v1.T)
print(u1)
print(np.dot(m1,u1.T))  # not exact but not very far

print('------')

# eigen

t = np.array([[2,0,0],[0,1,0],[0,0,2]])
w, v = linalg.eig(t)   
print(w)  # eig  values with repetition
print(v)  # normalized eig vectors
```

```
 # non linear regression : 

import numpy as np
from scipy.optimize import least_squares 
import matplotlib.pyplot as plt


def model(x,t) :
  return x[0]*np.log(x[1]*t+x[2])   # our model function

def target(x,t,y) :
  return model(x,t) - y     # setting it equel to y for our y values

def jac(x,t,y) :
  j = np.empty((t.size,x.size))        # the jacobian row with respect to the parameters x multiplied size of t times
  j[:,0] = x[0]
  j[:,1] = (x[0]*t)/(x[1]*t+x[2])
  j[:,2] = x[0]/(x[1]*t+x[2])
  return j

t = np.array([1, 1.1, 1.45, 2, 3.3, 3.73, 4.1, 4.76, 5])   # t sample values
y = np.array([0.5, 0.7, 0.4, 1.1, 0.9, 1.12, 1.34, 1.5, 1.2])  # y sample values
x0 = np.array([1,2,1.2])    # initial guess for the regression

func = least_squares(target,x0,jac=jac,args=(t,y))  # the regression takes all the above parameters.
print(func.x)  # the x results

t1 = np.linspace(1,6,50)
y1 = model(func.x, t1)
plt.plot(t,y,'o', label='data')
plt.plot(t1,y1,'-', label='fitted model')
plt.legend()
plt.xlabel('t',fontsize=15)
plt.ylabel('y',fontsize=15)
```

```
# different algorithms  :
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA as pca

data = datasets.load_linnerud()
algo1 = pca(n_components=2)  
algo1.fit(data.data)   
data1 = algo1.transform(data.data)
print(data1)
plt.scatter(data1[:,0],data1[:,1])


from sklearn.decomposition import TruncatedSVD as svd

algo2 = svd(n_components=2)
algo2.fit(data.data)
data2 = algo2.transform(data.data)
#print(data2)
#plt.scatter(data2[:,0],data2[:,1])


from sklearn.manifold import TSNE as tsne

data3 = tsne().fit_transform(data.data)
#print(data3)
#plt.scatter(data3[:,0],data3[:,1])

```

```
 # clustering :

import numpy as np 
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN, KMeans, MeanShift

np.random.seed(76)
data = 2*np.random.random(size=(20,2))  # some random numbers from 0 to 2
print(data)

#c = DBSCAN(eps=0.4,min_samples=4,metric='euclidean').fit(data).labels_  # can chose max dist, min points in cluster, metric type and more
c = KMeans(n_clusters=4).fit(data).labels_  # can chose number of clusters here
#c = MeanShift(bandwidth=0.6).fit(data).labels_
print(c) # the labels
plt.scatter(data[:,0],data[:,1],c=c)

```
